```python
# import pandas as pd

df = pd.read_csv('kg.csv', low_memory=False)

# See all relation types that involve our 3 node types
mask = (
    df["x_type"].isin(["disease", "effect/phenotype", "drug"]) &
    df["y_type"].isin(["disease", "effect/phenotype", "drug"])
)
subset = df[mask]

print(subset.groupby(["x_type", "relation", "y_type"]).size().reset_index(name="count").to_string())
```

```python
print(subset[["relation", "display_relation"]].drop_duplicates().to_string())
```
